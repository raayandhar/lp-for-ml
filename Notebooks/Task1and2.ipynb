```python
# Plot for 4.1 task 1
from math import ceil
percent_samples = [0.1 * i for i in range(1, 11)]
syn_perf = []
mnist_perf = []
syn_N = len(syn_data['trainX'])
mnist_N = len(mnist_data['trainX'])
for percent in percent_samples:
    syn_ind = np.random.choice(syn_N, ceil(syn_N * percent), replace=False)
    mnist_ind = np.random.choice(mnist_N, ceil(mnist_N * percent), replace=False)
    syn_class = Task1and2.MyClassifier(3)
    syn_class.train(syn_data['trainX'][syn_ind], syn_data['trainY'][syn_ind])
    mnist_class = Task1and2.MyClassifier(3)
    mnist_class.train(mnist_data['trainX'][mnist_ind], mnist_data['trainY'][mnist_ind])
    syn_perf.append(syn_class.evaluate(syn_data['testX'], syn_data['testY']))
    mnist_perf.append(mnist_class.evaluate(mnist_data['testX'], mnist_data['testY']))

plt.plot(percent_samples, syn_perf, label='synthetic')
plt.plot(percent_samples, mnist_perf, label='MNIST')
plt.xlabel('Percentage of input randomly sampled')
plt.ylabel('Accuracy')
plt.legend()
```
